# Numerical Integration: Trapezoidal Rule

Given a continous integrable function $f(x)$ in the interval $[a, b]$, we can make an approximation to the integral $\int_a^b f(x) dx$ by interpolating $f(x)$ in a finite set of points within the interval and integrating the interpolant instead of $f(x)$. 

The simplest interpolant for $f(x)$ is a linear polinomial, passing through the extreme points $(a, f(a))$, $(b, f(b))$. Such a simple rule is called the *trapezoidal rule*. Therefore:

$$
\int_a^b f(x) dx \approx h\cdot \left(\frac{f(a) + f(b)}{2}\right),
$$

where $h = b - a$, as preparation for the comming *composite trapezoidal rule*.

In order to ilustrate what numerical integration is about, let's define a concrete function, say $f(x) = e^{-x}$, and make the trapezoidal approximation in the interval $[0, 2]$:

```python
%matplotlib inline
```

```python
import numpy as np
from matplotlib import pyplot as plt


x = np.linspace(0, 2, 50)
y = np.exp(-x)

a, b = x[0], x[-1]
f_a, f_b = y[0], y[-1]

plt.plot(x, y, [a, b], [f_a, f_b], a, f_a, 'or', b, f_b, 'or')
plt.grid()
plt.legend(['$e^{-x}$', 'Linear approximation'])
```

As you can see in the image above, this is not a very good approximation, but consider what happens when we divide the interval in many subintervals of the same lenght. Suppose we do so, and divide the subinterval in 4, 8 and 16 subintervals:

```python
#This is a *lambda expression*, an easy way to define a short function.
f = lambda x: np.exp(-x)

N = (4, 8, 16)

xnodes = [np.linspace(0, 2, n + 1) for n in N]
ynodes = [f(node) for node in xnodes]

plots = []

for i, n in enumerate(N):
    plots.append(plt.plot(x, y, xnodes[i], ynodes[i], xnodes[i], ynodes[i], '.r'))
```

As you can see, with more nodes the approximation is better, as we would expect. The **composite trapezoidal rule** corresponds to applying the trapezoidal rule to each subinterval. Let's calculate the integral for the three cases above:

```python
exact_value = 1 - np.exp(-2)

header = 'Approximation, Error'

print header

for i, n in enumerate(N):
    h = 2. / n
    idx = np.arange(1, n)
    
    trap_approx = 0.5 * h * sum(ynodes[i][idx - 1] + ynodes[i][idx])
    
    line = '%0.4f, %0.4f' % (trap_approx, abs(exact_value - trap_approx))
    
    print line.rjust(len(header) + 1)
    

print ""
print 'Exact value:', '%0.4f' % exact_value
```

Don't worry about the last sentences in the loop, this is just fancy stuff to make clearer the printed data. What is important here is that the error is decreasing in a rather predescible way. Maybe we cannot say yet what is the rule to approximate the integral, but we could obtain insight into the answer by ploting the error for different nodes as we did for the derivatives in the previous notebooks. Let's do so to see what happens:

```python
N = np.arange(5, 51, 5)

errors = []

for n in N:
    x = np.linspace(0, 2, n + 1)
    y = f(x)
    h = 2. / n
    idx = np.arange(1, n)
    trap_approx = 0.5 * h * sum(y[idx - 1] + y[idx])
    errors.append(abs(trap_approx - exact_value))
    
h = 2. / N

plt.title('Error distribution in the trapezoidal rule')
plt.grid()
plt.xlabel('Nodes separation')
plt.ylabel('Absolute error')
plt.plot(h, errors, 'ro')
```

Our algorithm's error is following a linear rule!, so that we ought to prove that the composite trapezoidal rule is of order $O(h)$.

Let's calculate the least squares line interpolating the data as we did previously for the derivative.

```python
from scipy.linalg import lstsq


errors = np.array(errors)

model = np.ones((len(errors), 2))
model[:,1] = h

(a, b), res, rank, sv =  lstsq(model, errors)

print (a, b)
```

Recall that *lstsq* output contains the model coefficients, and three other values not very useful for us. In case you wonder what they are go take a look into a numerical analysis book (*hint:* *sv* stands for *singular values*)

```python
plt.title('Errors distribution')
plt.grid()
plt.xlabel('Nodes separation')
plt.ylabel('Absolute error')
plt.plot(h,a + b * h, h, errors, 'or')
plt.legend(['Least squares approx'], loc='upper left')
```

Finnally, note that in our algorithm, each inner node is counted twice. So, our implementation is slightly inneficient, as we are doing roughly two times the minimum number of sums. A slightly better approach would be to multiply each inner node by two and then sum them all (multiplying by two for a computer is as easy as multiplying by ten is for you).

However this rule is so simple that it should not make any noticeable difference.
