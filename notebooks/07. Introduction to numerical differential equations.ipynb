# Introduction to numerical differential equations

In this section we are going to solve ordinary differential equations. Given an equation of the form:

$$
\begin{aligned}
x'(t) &= f(x, t),\\
x(0) &= x_0,     
\end{aligned}
$$

$t \geq 0$, the *fundamental theorem of ordinary differential equations* asserts that, given *f* continuous in a neighborhood of $(x_0, 0)$, and provided it satisfies a *lipschitz condition* of the form

$$
|f(x, t) - f(y, t)| \leq k |x - y|,
$$

where $|x - y| < \delta$, and $k < 1$ is a constant, then it can be shown that there exists a unique solution to the equation in a neighborhood of $(x_0, 0)$. (ref?)

Suppose that we expand this solution in Taylor series around 0:

$$
x(h) = x(0) + x'(0)\cdot h + \frac{x''(\xi)}{2} h^2,
$$

for some $\xi \in (0, h)$. As $x(t)$ is a solution of the equation, we can use the previous expansion in order to obtain the following *recursive formula*:

$$
x(h) \approx x_0 + f(x_0, 0)\cdot h
$$

This is the celebrated *Euler formula*. As you can see from this equations, this is a *second order* method, provided we can guarantee a bounded second derivative for the solution.

## Phase portrait

Suppose we want to solve the ODE given by

$$
\begin{aligned}
x'(t) &= x(t),\\
x(0) &= x_0.
\end{aligned}
$$

As it's well known, The analytic solution is $x(t) = x_0 e^t$, which can be ploted for different values of $x_0$.

```python
import numpy as np
import sympy as sp
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
x0 = np.arange(7)
t = np.linspace(0, 2, 50)

sols = [x * np.exp(t) for x in x0]
```

```python
plt.hold(True)
for sol in sols:
    plt.plot(t, sol)
plt.hold(False)
```

Phase portrait:

```python
T, X = np.meshgrid(t, np.linspace(0, 45, 50))
U, V = np.ones(T.shape), X
plt.streamplot(T, X, U, V)
```

```python
plt.quiver(T, X, U, V)
```
